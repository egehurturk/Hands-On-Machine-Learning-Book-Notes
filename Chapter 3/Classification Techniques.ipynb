# MNIST

MNIST Dataset is a set of 70,000 small images of digits handwritten by high school students. This set is often called as the 'Hello World' of Machine Learning.

Scikit-Learn provides many helper functions to download popular dataset. MNIST is one of them:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

## MNIST Data Structure:

Datasets loaded by Scikit-learn generally have a similar structure:
* `DESCR` key describing the dataset
* A `data` key containing an array with one row per instance and one column per feature (x)
* A `target` key containing the labels (y)

```python
X, y = mnist['data'], mnist['target']
X.shape
```

```python
y.shape
```

So, we can interpret that we have 70000 examples, $m=70,000$, and 784 features, $n=784$. (28x28 images)

```python
import matplotlib as mpl

some_digit = X[0]
some_digit_image = some_digit.reshape(28,28)

plt.imshow(some_digit_image, cmap='binary')
plt.axis('off')
```

```python
y[0]
```

The image looks like 5, and that's what the label tells us.

Since the y value is a string, we should convert that into numbers:

```python
y = y.astype(np.uint8)
```

```python
# Some helper functions to plot image:
def plot_digit(data):
    image = data.reshape(28, 28)
    plt.imshow(image, cmap = mpl.cm.binary,
               interpolation="nearest")
    plt.axis("off")
    
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")


plt.figure(figsize=(9,9))
example_images = X[:100]
plot_digits(example_images, images_per_row=10)
plt.show()
```

We should always create a test set and set is aside before inspecting the data closely. 

A nice thing about MNIST Datasets is that they are already split into a training set (the first 60,000 images) and a test set (last 10,000 images). Shuffled dataset is good 

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

# Training a Binary Classifier

To simplify the problem, we are going to implement a binary classifier - which classifies 5 and non-5s.

```python
y_train_5 = (y_train==5) # True for 5s, False for non-5s
y_test_5 = (y_test == 5)
```

## Model Selection

Lets select the $Stochastic\ Gradient\ Descent\ Classifier$ since it is capable of handling large datasets efficiently (one instance per epoch)

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
# Predicting
sgd_clf.predict([some_digit])
```

The classifier guessed it right! Let's look at the performance of this classifier.

# Performance Measures

Evaluating a classifier is always trickier than a regressor.

## Measuring Accuracy Using Cross-Validation

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_5, cv=4, scoring='accuracy')
```

Wow, over 90% accuracy! This is simply because only about 10% of the images are 5s, so if you have always guess that an image is not a 5, you will be right about 90% of time.
\
This shows that **accuracy is not generally preferred when you are dealing with skewed datasets**.

## Confusion Matrix

The general idea is to count the number of times instances of class A are classified as class B. For example, to know the number of times the classifier confused images of 5s with 3s, you would look in the fifth row and third column of the confusion matrix.
\
\
To compute confusion matrix, we should have a set of predictions. We don't want to use the test set, so we will use the `cross_val_predict()` function of Scikit-learn. `cross_val_score()` performs K-fold cross validation, but instead of returning the scores, it returns the predictions that the classifier predicts on each test fold.

```python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

Now, we are ready to get the confusion matrix using the `confusion_matrix()` function. \
**Parameters are**:
* The target class  (`y_train_5`)
* The predicted class (`y_train_pred`)

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5, y_train_pred)
```

Each row in a confusion matrix representes an **actual class**, while each column represents the **predicted class**. The first row of the matrix considers non-5 images (the **negative class**). 53,892 of them were correctly classified as non-5s (they are called the **true negatives**), while the remaning 687 were wrongly classified as 5 (**false positive**).
\
\
The second row of the matrix considers the images of 5s (the **positive class**). 1891 of them were wrongly classified as non-5s (**false negatives**), while the remaining 3530 were correctly classified as 5s. (**true positives**)
\
\
A perfect classifier would have only TP and TN.

--------------------

The confusion matrix gives us a lot of information, but there are others to consider. One of them is the *accuracy of positive predictions*, which is called the **precision** of the classifier$$Precision: \\ precision = \frac{TP}{TP+FP}$$ where TP is the number of true positives, and FP is the number of false positives.

Precision is commonly used with another metric called **recall**, also called the *sensivity* or the *true positive rate*. This is the **ratio of positive instances that are correctly detected**.$$Recall:\\recall = \frac{TP}{TP+FN}$$

## Precision and Recall

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y_train_5, y_train_pred)
```

```python
recall_score(y_train_5, y_train_pred)
```

When it claims that an image is 5, it is correct only 83.8% of the time. It only detects 65% of the 5s.

It is often convenient to combine these 2 metrics into a single one, called **f1 score**. $$F_{1} = \frac{2}{\frac{1}{precision} + \frac{1}{recall}} = 2\times \frac{precision\times recall}{precision+recall} = \frac{TP}{TP+\frac{FN+FP}{2}}$$

To compute the f1 score, simply call the `f1_score()` function.

```python
from sklearn.metrics import f1_score
f1_score(y_train_5, y_train_pred)
```

Increasing precision reduces recall, and vice versa. This is called the *Precision Recall Tradeoff*

## Precision/Recall Trade-off

A way to select good precision/recall tradeoff is to plot the precision directly against recall.

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3,
                             method="decision_function")

from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)

def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.plot([0.4368, 0.4368], [0., 0.9], "r:")
plt.plot([0.0, 0.4368], [0.9, 0.9], "r:")
plt.plot([0.4368], [0.9], "ro")
plt.show()
```

We can see that the precision starts to fall sharply around 80% recall. You will probably want to select a precision/recall trade-off just before that drop, around 60% recall. 

## The ROC Curve

The *reciever operating characteristics* (ROC) curve is another common tool used with **binary classifiers**. It plots Recall against False Positive Rate.
\

$$ ROC= \frac{precision}{1-precision}$$To plot the ROC curve, you should use the `roc_curve()` function to co,pute the TPR and FPR for various threshold values:

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
    plt.axis([0, 1, 0, 1])                                    # Not shown in the book
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) # Not shown
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    # Not shown
    plt.grid(True)                                            # Not shown

plt.figure(figsize=(8, 6))                         # Not shown
plot_roc_curve(fpr, tpr)
plt.plot([4.837e-3, 4.837e-3], [0., 0.4368], "r:") # Not shown
plt.plot([0.0, 4.837e-3], [0.4368, 0.4368], "r:")  # Not shown
plt.plot([4.837e-3], [0.4368], "ro")               # Not shown                       # Not shown
plt.show()
```

A good classifier should always stay as far as possible from that dotted line.


One way to compare classifiers is to measure the *area under the curve* (AUC). A perfect classifier would have a ROC AUC equal to 1, whereas a random classifier would have a ROC AUC equal to 0.5. Scikit-learn provides a function to compute ROC AUC:

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_scores)
```

# Multiclass Classification:

Some algorithms (Logistic Regression, Random Forest, Naive Bayes) are capable of handling multiple classes natively. Others (SGD, SVM) are strictly bunary classifiers. However, there aer various strategies that you can use to perform multiclass classification with multiple binary classifiers.

One way to create a system that can classify the digit images into 10 classes (from 0 to 9) is to train 10 binary classifiers for each digit: 0 classifiers, 1 classifiers, etc. Then, you would select the highest probability. This is called **One Versus Rest** strategy.

Another strategy is to train a binary classifier for every pair of digits: one to distinguish 0-1, another to distinguish 0-2, etc. This is called the **One Versus One** strategy. If there are $N$ classifiers, you need to train $N \times \frac{(N-1)}{2}$ classifiers. Some algorithms (SVMClassifier) scale poorly with the size of the training set. For these algorithms, OvO is preferred because it is faster to train many classifiers on small trainng sets than to trian few classifiers on large training set.
